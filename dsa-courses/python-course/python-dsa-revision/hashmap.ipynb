Hash table or hash map is a type of data structure that maps keys to its value pairs.

## Creating a Dictionary

```python
my_dict = {'Leon': '001', 'Daniel': '002', 'Nicholas': '003'}
print(my_dict)
type(my_dict)
```

```python
new_dict = dict()
new_dict = dict(Leon='001', Daniel='002')
print(new_dict)
```

## Nested Dictionaries

- Dictionaries that lie within other dictionaries

```python
emp_details = {'Employee': {'Dave':{'ID':'001','Salary':'2000','Designation':'Team Lead'}, 
                            'Ava':{'ID':'002','Salary':'1000','Designation':'Associate'}}}
print(emp_details)
```

## Performing Operations on Hash Tables
- Accessing Items
- Updating Values
- Deleting Entries

#### Accessing Items

```python
# Accessing Items
my_dict = {'Leon': '001', 'Daniel': '002', 'Nicholas': '003'}
my_dict['Leon']
```

```python
print(my_dict.keys())
print(my_dict.values())
print(my_dict.get('Leon'))
```

```python
# Keys
for x in my_dict:
    print(x)

# Values
for x in my_dict.values():
    print(x)
```

```python
for x, y in my_dict.items():
    print(x, y)
```

#### Updating Values
- Dictionaries are mutable data types and you can update them as and when required.

```python
my_dict = {'Leon': '001', 'Daniel': '002', 'Nicholas': '003'}
my_dict['Leon'] = '004'
my_dict['Jie Wei'] = '003'
print(my_dict)
```

#### Deleting Entries

```python
my_dict = {'Leon': '001', 'Daniel': '002', 'Nicholas': '003', 'Joe':'003'}

# pop() function
my_dict.pop('Leon')
print(my_dict)
```

```python
# remove the last item in the dictionary
my_dict.popitem()
```

```python
del my_dict['Nicholas']
print(my_dict)
```
